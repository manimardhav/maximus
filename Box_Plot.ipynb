```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Create a sample dataset with some outliers
np.random.seed(10)
data = pd.DataFrame({
    'value': np.concatenate([np.random.normal(0, 20, 1000), np.random.normal(50, 1, 20)])  # Add outliers at the high end
})

# Calculate the IQR for box plot-based outlier detection
Q1 = data['value'].quantile(0.25)
Q3 = data['value'].quantile(0.75)
IQR = Q3 - Q1

# Define outlier thresholds
lower_bound = Q1 - 2.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify outliers
outliers = data[(data['value'] < lower_bound) | (data['value'] > upper_bound)]

# Print outliers
print("Outliers:\n", outliers)

# Create the boxplot
plt.figure(figsize=(10, 6))
sns.boxplot(data['value'], color='lightblue', flierprops=dict(marker='o', color='red', markersize=7))

# Customize the title and labels
plt.title('Boxplot with Outliers')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.grid(True, axis='y', linestyle='--', alpha=0.7)

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

datf = pd.DataFrame({
    "Season 1": [3, 67, 5, 5, 56, 3],
    "Season 2": [12, 23, 14, 34, 34, None]
})

# Convert the DataFrame to a long format
datf_melted = datf.melt(var_name='Season', value_name='Value')

# Create the boxplot with seaborn
plt.figure(figsize=(10, 6))
P = sns.boxplot(data=datf_melted, x='Season', y='Value', palette='pastel')

# Customize the title and labels
P.set_title('Boxplot of Values by Season')
P.set_xlabel('Season')
P.set_ylabel('Value')

plt.grid(True, axis='y', linestyle='--', alpha=0.7)

# Show the plot
plt.show()

```

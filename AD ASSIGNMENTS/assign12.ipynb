```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = r'C:\ML\Day_12_banking_data.csv'
df = pd.read_csv(file_path)

# Task 1: Plot the total sum of Transaction_Amount per Account_Type using a bar plot
total_transaction_amount_per_account_type = df.groupby('Account_Type')['Transaction_Amount'].sum()

plt.figure(figsize=(10, 6))
total_transaction_amount_per_account_type.plot(kind='bar', color='skyblue')
plt.title('Total Transaction Amount per Account Type')
plt.xlabel('Account Type')
plt.ylabel('Total Transaction Amount')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()

# Task 2: Create a pie chart to show the percentage of transactions per Branch
transactions_per_branch = df['Branch'].value_counts()

plt.figure(figsize=(8, 8))
transactions_per_branch.plot(kind='pie', autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)
plt.title('Percentage of Transactions per Branch')
plt.ylabel('')
plt.show()
```

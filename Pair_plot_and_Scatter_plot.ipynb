```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

#Expand dataset
data={
    'Temparature': [25,28,32,35,38,40,42,45,30,33,36,37],
    'Ice_Cream_Sales': [200,250,300,350,400,420,430,450,280,310,360,370],
    'Advertising_Budget': [50,55,60,70,75,80,85,90,60,65,70,75],
    'Customer_Count': [80,100,120,140,150,160,170,180,110,130,140,145],
}

#Convert to Dataframe
df=pd.DataFrame(data)

#Pair plot
sns.pairplot(df,diag_kind='kde',markers="o",plot_kws={'alpha':0.7})
plt.suptitle("Pair Plot of Ice Cream Sales Data", y=1.02, fontsize=16)
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

#Expand dataset
data={
    'Temparature': [25,28,32,35,38,40,42,45,30,33,36,37],
    'Humidity': [200,250,300,350,400,420,430,450,280,310,360,370],
    'Wind Speed': [50,55,60,70,75,80,85,90,60,65,70,75],
    'Pressure': [80,100,120,140,150,160,170,180,110,130,140,145],
}

#Convert to Dataframe
df=pd.DataFrame(data)

#Scatter Plot
plt.figure(figsize=(8,6))
plt.scatter(df['Temparature'],df['Humidity'],color='blue',label='Temp vs Humidity')

#Adding Labels and title
plt.title('Temperature vs Humidity',fontsize=14)
plt.xlabel('Temperature (*c)',fontsize=12)
plt.ylabel('Humidity')
plt.legend()
plt.grid(True)
```
